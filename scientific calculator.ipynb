```python
import tkinter as tk
import math

def evaluate_expression():
    try:
        result.set(eval(entry.get()))
    except Exception as e:
        result.set("Error")

def clear():
    entry.delete(0, tk.END)
    result.set("")

def append_to_expression(value):
    current_expression = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current_expression + value)

def calculate_sqrt():
    try:
        value = float(entry.get())
        result.set(math.sqrt(value))
    except Exception as e:
        result.set("Error")

def calculate_trig_function(func):
    try:
        angle = float(entry.get())
        if func == "sin":
            result.set(math.sin(math.radians(angle)))
        elif func == "cos":
            result.set(math.cos(math.radians(angle)))
        elif func == "tan":
            result.set(math.tan(math.radians(angle)))
    except Exception as e:
        result.set("Error")

def calculate_logarithm(base):
    try:
        value = float(entry.get())
        if base == "log":
            result.set(math.log10(value))
        elif base == "ln":
            result.set(math.log(value))
    except Exception as e:
        result.set("Error")

# Create the main window
root = tk.Tk()
root.title("Scientific Calculator")

# Entry widget for input
entry = tk.Entry(root, width=30)
entry.grid(row=0, column=0, columnspan=5)

# Result display
result = tk.StringVar()
result_label = tk.Label(root, textvariable=result)
result_label.grid(row=1, column=0, columnspan=5)

# Calculator buttons
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+',
    'sqrt', 'sin', 'cos', 'tan', 'log', 'ln', '^', '(', ')', 'C'
]

row_val = 2
col_val = 0

for button in buttons:
    if button == '=':
        tk.Button(root, text=button, padx=20, pady=20, command=evaluate_expression).grid(row=row_val, column=col_val)
    elif button == 'C':
        tk.Button(root, text=button, padx=20, pady=20, command=clear).grid(row=row_val, column=col_val)
    elif button == 'sqrt':
        tk.Button(root, text=button, padx=20, pady=20, command=calculate_sqrt).grid(row=row_val, column=col_val)
    elif button in ('sin', 'cos', 'tan'):
        tk.Button(root, text=button, padx=20, pady=20, command=lambda b=button: calculate_trig_function(b)).grid(row=row_val, column=col_val)
    elif button in ('log', 'ln'):
        tk.Button(root, text=button, padx=20, pady=20, command=lambda b=button: calculate_logarithm(b)).grid(row=row_val, column=col_val)
    else:
        tk.Button(root, text=button, padx=20, pady=20, command=lambda b=button: append_to_expression(b)).grid(row=row_val, column=col_val)
    
    col_val += 1
    if col_val > 4:
        col_val = 0
        row_val += 1

# Start the main loop
root.mainloop()
```
